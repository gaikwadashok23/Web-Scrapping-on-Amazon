```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
driver = webdriver.Chrome()
driver.get("http://www.python.org")
assert "Python" in driver.title
elem = driver.find_element(By.NAME, "q")
elem.clear()
elem.send_keys("pycon")
elem.send_keys(Keys.RETURN)
assert "No results found." not in driver.page_source
time.sleep(12)
driver.close()
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
driver = webdriver.Chrome()
query="laptop"
file=0
for i in range(2,11):
    driver.get(f"https://www.amazon.in/s?k={query}&page={i}&crid=1I4MOMTVBBY8T&sprefix=laptop%2Caps%2C245&ref=nb_sb_noss_2")
    elems=driver.find_elements(By.CLASS_NAME, "puis-card-container")
    print(f"{len(elems)} items found")
    for elem in elems:
        d=elem.get_attribute("outerHTML")
        with open(f"selenium_file/{query}_{file}.html", "w" ,encoding="utf=8") as f:
            f.write(d)
            file+=1
        #print(elem.text)
    time.sleep(4)
driver.close()
```

```python
from bs4 import BeautifulSoup
import os
import pandas as pd

# Sample dictionary (d) provided in the original code
d = {'title': [], 'price': [], 'link': []}

# Iterate through files in the "selenium_file" directory
for file in os.listdir("selenium_file"):
    # Skip hidden files and directories
    try:
        if file.startswith('.'):
           continue
    
        with open(f"selenium_file/{file}") as f:
           html_doc = f.read()
    
    # Parse the HTML document
        soup = BeautifulSoup(html_doc, "html.parser")
    
    # Extract the title from the 'h2' tag
        t = soup.find("h2")
        if t:
           title = t.get_text()
           l=t.find("a")
           link="https://amazon.in/" + l['href']
           p=soup.find("span", attrs={"class": 'a-price-whole'})
           price=p.get_text()
           d['title'].append(title)
           d['price'].append(price)
           d['link'].append(link)
           #print(title,link,price)
    except Exception as e:
        print(e)
      
df=pd.DataFrame(data=d)
df.to_csv("data.csv")
```
